```python
import pandas as pd
import h3

# Load the CSV file
file_path = r"Z:\Thesis\Data\test\DustCast\UA\DC_v0063\ua_row_level_predictions_2023_v0063.csv"
data = pd.read_csv(file_path)

# Function to calculate H3 index for multiple resolutions
def calculate_h3_indexes(lat, lon, resolutions):
    return {f"h3_res_{res}": h3.geo_to_h3(lat, lon, res) for res in resolutions}

# Specify the H3 resolutions
resolutions = [3, 4]

# Apply the function and expand the H3 indexes into the dataframe
h3_columns = data.apply(
    lambda row: calculate_h3_indexes(row['lat'], row['lon'], resolutions),
    axis=1
)
h3_df = pd.DataFrame(h3_columns.tolist())

# Concatenate the original data with the H3 index columns
data_with_h3 = pd.concat([data, h3_df], axis=1)

# Save the updated dataframe to a new CSV file
output_path = r"Z:\Thesis\Data\test\DustCast\UA\DC_v0063\ua_row_level_predictions_2023_v0063_H3res4.csv"   #_H3res4
data_with_h3.to_csv(output_path, index=False)

print(f"H3-indexed data saved to: {output_path}")
```
