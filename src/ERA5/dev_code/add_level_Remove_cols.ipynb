```python
# modify Parquet files
# Add pressure_level column with 1013mb entry for sfc data
# remove extra number and expver columns if they exist

# Import libraries
import os
import xarray as xr
import pandas as pd
from tqdm import tqdm

# Define the root directory where the parquet files are saved
root_dir = r"Z:\Thesis\Data\Met\ERA5_parquet"

# Define the years to process
years = [str(year) for year in range(1980, 1988)]

# Define the list of countries
countries = [
    "Bahrain",
    "Kuwait",
    "Oman",
    "Qatar",
    "Saudi Arabia",
    "United Arab Emirates",
    "Yemen"
]

# Define the subfolders to process
subfolders =["pressure", "surface"]

# Collect all Parquet files into a list
parquet_files = []

# Iterate through the years
for year in years:
    year_dir = os.path.join(root_dir, year)
    #iterate through the countries
    for country in countries:
        country_dir = os.path.join(year_dir, country)
        # Iterate through the subfolders
        for subfolder in subfolders:
            subfolder_dir = os.path.join(country_dir, subfolder)
            # Check if the subfolder exists
            if not os.path.exists(subfolder_dir):
                continue
            # Get a list of Parquet files in the subfolder
            files_in_subfolder = [
                os.path.join(subfolder_dir, f) for f in os.listdir(subfolder_dir) if f.endswith('.parquet')
            ]
            # Add the files to the list
            parquet_files.extend(files_in_subfolder)

# Process the files with tqdm progress bar
for parquet_path in tqdm(parquet_files, desc="Processing Parquet files"):
    # Print file being processed
    tqdm.write(f"Processing file: {parquet_path}")
    # Read the parqeut file using pandas
    try:
        df = pd.read_parquet(parquet_path)
    except Exception as e:
        tqdm.write(f"Error reading {parquet_path}: {e}")
        continue

    # Remove 'number' and 'expver' columns if they exist
    columns_to_remove = ['number', 'expver']
    df = df.drop(columns=[col for col in columns_to_remove if col in df.columns], errors='ignore')

    # Check if the file is in a 'surface' subfolder to add 'pressure_level' column
    if os.path.normpath('surface') in os.path.normpath(parquet_path).split(os.sep):
        df['pressure_level'] = 1013.0
        # Ensure 'pressure level is the second column after 'valid_time' for consistency
        cols = df.columns.tolist()
        cols.insert(1, cols.pop(cols.index('pressure_level')))
        df = df[cols]

    # write the modified DataFrame back to the same Parquet file
    try:
        df.to_parquet(parquet_path, index=False)
        tqdm.write(f"Successfully updated {parquet_path}")
    except Exception as e:
        tqdm.write(f"Error writing Parquet file for {parquet_path}: {e}")
        continue
```
