```python
#SFC

import pandas as pd
import os

# Define file paths
input_file = r"Z:\Thesis\Data\test\DustCast\SFC\DC_v0063\sfc_row_level_predictions_2023_v0063_H3res4.csv"
output_file = r"Z:\Thesis\Data\test\DustCast\SFC\DC_v0063\sfc_row_level_predictions_2023_v0063_H3res4_with_MSE.csv"

# Read the CSV file
df = pd.read_csv(input_file)

# Check that required columns exist
if "DUSMASS_mean" not in df.columns or "ensemble_predictions" not in df.columns:
    raise ValueError("The input file must contain the columns 'DUSMASS_mean' and 'ensemble_predictions'.")

# Calculate the Mean Squared Error for each row (i.e. squared error)
df["MSE"] = (df["DUSMASS_mean"] - df["ensemble_predictions"]) ** 2

# Save the updated DataFrame to a new CSV file
df.to_csv(output_file, index=False)
print(f"Updated file saved to {output_file}")
```
