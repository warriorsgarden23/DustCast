```python
#Monthly conversion of DUSMAS_Mean, DustCast_Predictions, and Difference error

import pandas as pd
import numpy as np

# File paths
input_path = r"Z:\Thesis\Data\test\DustCast\SFC\DC_v0063\sfc_row_level_predictions_2023_v0063_H3res4.csv"
output_path = r"Z:\Thesis\Data\test\DustCast\SFC\DC_v0063\monthly_µg_RMSE.csv"

# Read the CSV file
df = pd.read_csv(input_path)

# List of columns to extract (adjust if the CSV has different column names)
cols_to_extract = [
    'lon', 'lat', 'time', 
    'DUSMASS_mean', 'Month', 'ensemble_predictions', 
    'Country', 'h3_res_3', 'h3_res_4'
]

# Extract the columns
df_extracted = df[cols_to_extract].copy()

# Convert from kg/m^3 to µg/m^3: 1 kg = 1e9 µg
df_extracted['DUSMASS_mean_ugm3'] = df_extracted['DUSMASS_mean'] * 1e9
df_extracted['ensemble_predictions_ugm3'] = df_extracted['ensemble_predictions'] * 1e9

# Compute the error per row (difference in µg/m^3)
# Note: For each row, error = (observation - prediction). 
# Since RMSE is the square root of the mean of squared errors across rows,
# the per-row absolute error is |observation - prediction|.
df_extracted['error'] = df_extracted['DUSMASS_mean_ugm3'] - df_extracted['ensemble_predictions_ugm3']

# Compute overall RMSE across all rows
rmse = np.sqrt(np.mean(df_extracted['error']**2))
print(f"Overall RMSE: {rmse:.3f} µg/m^3")

# Save the resulting dataframe to a new CSV file
df_extracted.to_csv(output_path, index=False)
```

```python
import pandas as pd
import numpy as np

# File paths
input_path = r"Z:\Thesis\Data\test\DustCast\SFC\DC_v0063\sfc_row_level_predictions_2023_v0063_H3res4.csv"
output_path = r"Z:\Thesis\Data\test\DustCast\SFC\DC_v0063\monthly_µg_RMSE.csv"

# Read the CSV file
df = pd.read_csv(input_path)

# List of columns to extract (adjust if the CSV has different column names)
cols_to_extract = [
    'lon', 'lat', 'time', 
    'DUSMASS_mean', 'Month', 'lag_1', 'ensemble_predictions', 
    'Country', 'h3_res_3', 'h3_res_4'
]

# Extract the columns
df_extracted = df[cols_to_extract].copy()

# Convert from kg/m^3 to µg/m^3: 1 kg = 1e9 µg
df_extracted['DUSMASS_mean_µgm3'] = df_extracted['DUSMASS_mean'] * 1e9
df_extracted['ensemble_predictions_µgm3'] = df_extracted['ensemble_predictions'] * 1e9

# Compute the difference per row (observation - prediction)
df_extracted['Diff_error'] = df_extracted['DUSMASS_mean_µgm3'] - df_extracted['ensemble_predictions_µgm3']

# Compute per row metrics:
# MSE: squared error per row
df_extracted['MSE'] = df_extracted['Diff_error'] ** 2
# MAE: absolute error per row
df_extracted['MAE'] = df_extracted['Diff_error'].abs()
# RMSE: square root of the squared error per row (i.e., absolute error)
df_extracted['RMSE'] = np.sqrt(df_extracted['MSE'])

# (Optional) Print overall RMSE computed across all rows for reference
overall_rmse = np.sqrt(np.mean(df_extracted['Diff_error'] ** 2))
print(f"Overall RMSE: {overall_rmse:.3f} µg/m^3")

# Save the resulting dataframe to a new CSV file
df_extracted.to_csv(output_path, index=False)
```
