```python

#UA as mg/m^2
import os
import pandas as pd

# List of input file paths
files = [
    r"Z:\Thesis\Data\test\DustCast\UA\DC_v0063\UA_aggregated_quarterly_µg_prediction_errors2.csv",
    r"Z:\Thesis\Data\test\DustCast\SFC\DC_v0063\SFC_aggregated_quarterly_µg_prediction_errors2.csv"
]

# List of quarters to process
quarters = ['DJF', 'MAM', 'JJA', 'SON']

for file_path in files:
    # Read the CSV file
    df = pd.read_csv(file_path)
    
    # For UA file, convert numeric aggregated columns from µg/m² to mg/m².
    if "UA" in file_path:
        conversion_map = {
            "agg_Mean_DUCMASS_mean_ugm3": 1/1000.0,
            "agg_Mean_ensemble_predictions_ugm3": 1/1000.0,
            "agg_Diff_error": 1/1000.0,
            "agg_MAE": 1/1000.0,
            "agg_RMSE": 1/1000.0,
            "agg_MSE": 1/1000000.0  # Because MSE is squared
        }
        for col, factor in conversion_map.items():
            if col in df.columns:
                df[col] = df[col] * factor
    
    # Get the directory of the current file and create a "Quarterly_aggs" folder inside it
    directory = os.path.dirname(file_path)
    output_dir = os.path.join(directory, "Quarterly_aggs")
    os.makedirs(output_dir, exist_ok=True)
    
    # Use the base file name (without .csv) for naming the output files
    base_filename = os.path.splitext(os.path.basename(file_path))[0]
    
    # Process each quarter
    for quarter in quarters:
        # Filter rows for the current quarter
        df_quarter = df[df["Quarter"] == quarter]
        
        # Define output file name: e.g. "UA_aggregated_quarterly_µg_prediction_errors2_DJF.csv"
        output_file = os.path.join(output_dir, f"{base_filename}_{quarter}.csv")
        
        # Save the filtered dataframe to CSV
        df_quarter.to_csv(output_file, index=False)
        print(f"Saved {output_file}")
```
