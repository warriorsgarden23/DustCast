```python
# Create SFC monthly files with pertinent data for MSE visualizations

import os
import pandas as pd

# File paths
file_path = r"Z:\Thesis\Data\test\DustCast\SFC\DC_v0063\sfc_row_level_predictions_2023_v0063_H3res4_with_MSE.csv"
output_dir = r"Z:\Thesis\Data\test\DustCast\SFC\DC_v0063\qgis_monthly_data\MSE"  # SFC

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Load the data
data = pd.read_csv(file_path)

# Specify the columns to keep
columns_to_keep = [
    'lon', 
    'lat', 
    'time', 
    'DUSMASS_mean', 
    'ensemble_predictions',  # Adjusted column name (if needed)
    'Country', 
    'h3_res_3', 
    'h3_res_4', 
    'MSE'
]

# Subset the DataFrame to only the desired columns
data = data[columns_to_keep]

# If 'Month' is not already a column, derive it from 'time'
if 'Month' not in data.columns:
    # Convert 'time' to datetime and extract the month as integer
    data['time'] = pd.to_datetime(data['time'])
    data['Month'] = data['time'].dt.month

# Group data by Month and save each group to a CSV file
for month, group in data.groupby('Month'):
    output_file = os.path.join(output_dir, f"DC_v0.0.6.3_sfc_{month}_MSE.csv")
    group.to_csv(output_file, index=False)

print(f"Monthly files have been successfully created and saved to {output_dir}.")
```

```python
#UA monthly files by level
import pandas as pd
import os

# File paths
input_file_path = r"Z:\Thesis\Data\test\DustCast\UA\UA_monthly_with_h3_indexes.csv"
output_directory = r"Z:\Thesis\Data\test\DustCast\UA\monthly_outputs"

# Ensure the output directory exists
os.makedirs(output_directory, exist_ok=True)

# Load the data
data = pd.read_csv(input_file_path)

# Unique combinations of Month and Pressure Level
for (month, level), group in data.groupby(['Month', 'Level']):
    # Create a subdirectory for each month
    month_dir = os.path.join(output_directory, month)
    os.makedirs(month_dir, exist_ok=True)
    
    # Save each pressure level file within the month directory
    output_file = os.path.join(month_dir, f"DC_v0.0.3.2_ua_{month}_{level}.csv")
    group.to_csv(output_file, index=False)

print(f"Data segregated by month and pressure level saved to: {output_directory}")
```
