```python
import pandas as pd
import pyarrow.parquet as pq

def read_and_search_parquet(file_path, column_name, search_value):
    """
    Reads the last 50 lines of a Parquet file, filters the data based on a column and search value,
    and returns the filtered DataFrame.
    """
    try:
        # Load the Parquet file into a DataFrame
        df = pd.read_parquet(file_path, engine='pyarrow')
        print(f"Data successfully loaded from: {file_path}")
        
        # Get the last 50 rows
        last_50_rows = df.tail(50)
        
        # Display the last 50 rows to verify data
        print("Last 50 rows of the Parquet file:")
        print(last_50_rows)
        
        # Filter the data for the specified column and value
        if column_name in last_50_rows.columns:
            filtered_rows = last_50_rows[last_50_rows[column_name] == search_value]
            print(f"Filtered rows where {column_name} == {search_value}:")
            print(filtered_rows)
            return filtered_rows
        else:
            print(f"Column '{column_name}' not found in the data.")
            return None
        
    except Exception as e:
        print(f"Error reading Parquet file: {e}")
        return None

# Define the path to the Parquet file
file_path = r"Z:\Thesis\Data\Met\ERA5_parquet_test\2017\Bahrain\surface\2017_Bahrain_surface_2m_dewpoint_temperature.parquet"

# Specify the column name and value to search
column_name = 'level'
search_value = 500  # Example: Looking for rows where 'level' == 500

# Read and search the Parquet file
filtered_df = read_and_search_parquet(file_path, column_name, search_value)

# Example analysis: display column names and data types
if filtered_df is not None:
    print("Columns and Data Types:")
    print(filtered_df.dtypes)
```
